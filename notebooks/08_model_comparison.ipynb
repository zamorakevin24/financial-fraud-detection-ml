```python
import pandas as pd
```

```python
results = {
    "Model": [
        "Logistic Regression",
        "Random Forest",
        "XGBoost",
        "Hybrid Model"
    ],
    
    "Precision": [
        0.79,
        0.94,
        0.81,
        0.94
    ],
    
    "Recall": [
        0.69,
        0.83,
        0.85,
        0.82
    ],
    
    "F1-score": [
        0.74,
        0.88,
        0.83,
        0.87
    ],
    
    "ROC-AUC": [
        0.92,
        0.95,
        0.97,
        0.957
    ]
}

df_results = pd.DataFrame(results)

df_results
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(df_results["Model"], df_results["Recall"])

plt.title("Fraud Detection Recall by Model")

plt.ylabel("Recall")

plt.xticks(rotation=30)

plt.show()
```
