```python
plt.figure(figsize=(8,4))
plt.hist(df["Time"], bins=50)
plt.title("Distribution of Time")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()
```

### Distribution of Transaction Time

The `Time` variable represents the number of seconds elapsed since the first transaction in the dataset. 

The histogram shows two main peaks, suggesting that the dataset contains transactions spanning approximately two days. The lower activity between the peaks likely corresponds to periods of reduced financial activity (e.g., nighttime hours).

Temporal patterns in transaction activity may be useful for detecting anomalies or unusual transaction behavior.

```python
fraud = df[df["Class"] == 1]
normal = df[df["Class"] == 0]

plt.figure(figsize=(10,5))

plt.hist(normal["Amount"], bins=50, alpha=0.5, label="Normal")
plt.hist(fraud["Amount"], bins=50, alpha=0.5, label="Fraud")

plt.legend()
plt.title("Transaction Amount: Fraud vs Normal")
plt.xlabel("Amount")
plt.ylabel("Frequency")

plt.show()
```

```python
import numpy as np

df["Amount_log"] = np.log1p(df["Amount"])
```

```python
fraud = df[df["Class"] == 1]
normal = df[df["Class"] == 0]

plt.figure(figsize=(10,5))

plt.hist(normal["Amount_log"], bins=50, alpha=0.5, label="Normal")
plt.hist(fraud["Amount_log"], bins=50, alpha=0.5, label="Fraud")

plt.legend()
plt.title("Transaction Amount (log scale): Fraud vs Normal")
plt.xlabel("Log Amount")
plt.ylabel("Frequency")

plt.show()
```

```python
corr = df.corr()

corr["Class"].sort_values(ascending=False)
```

```python
top_features = corr["Class"].drop("Class").abs().sort_values(ascending=False).head(10)

top_features.plot(kind="bar")
plt.title("Top Features Correlated with Fraud")
plt.ylabel("Correlation")
plt.show()
```
